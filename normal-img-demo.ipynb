```python
import os

for root, dirs, files in os.walk('/kaggle/input'):
    print(root)
    for f in files:
        print("   ", f)
```

```python
from ultralytics import YOLO
from IPython.display import Image as IPImg, display

model_surface = YOLO('/kaggle/input/datasets/kolaouk/trained-solar-models/surface_defect_best.pt')


TEST_IMAGE = '/kaggle/input/datasets/kolaouk/electrical-img/Electrical (10).jpg'

display(IPImg(TEST_IMAGE, width=400))

result = model_surface.predict(TEST_IMAGE, verbose=False)[0]
predicted = model_surface.names[int(result.probs.top1)]
confidence = float(result.probs.top1conf)
print(f"Predicted: {predicted} ({confidence:.2%})")
```
